```python
import pandas as pd

# Hardcoded crop mapping
crop_mapping = {
    'Arecanut': {'Duration': 'Long', 'Height': 'Tall', 'Structure': 'Tree', 'Structure_Numeric': 1, 'duration_numeric': 3, 'height_numeric': 3},
    'Avare': {'Duration': 'Low', 'Height': 'Medium', 'Structure': 'Shrub', 'Structure_Numeric': 2, 'duration_numeric': 1, 'height_numeric': 2},
    'Bajra': {'Duration': 'Low', 'Height': 'Short', 'Structure': 'Grass Crop', 'Structure_Numeric': 3, 'duration_numeric': 1, 'height_numeric': 1},
    'Banana': {'Duration': 'Long', 'Height': 'Tall', 'Structure': 'Tree', 'Structure_Numeric': 1, 'duration_numeric': 3, 'height_numeric': 3},
    'Coconut': {'Duration': 'Long', 'Height': 'Tall', 'Structure': 'Tree', 'Structure_Numeric': 1, 'duration_numeric': 3, 'height_numeric': 3},
    'Eucalyptus': {'Duration': 'Long', 'Height': 'Tall', 'Structure': 'Tree', 'Structure_Numeric': 1, 'duration_numeric': 3, 'height_numeric': 3},
    'Guava': {'Duration': 'Long', 'Height': 'Tall', 'Structure': 'Tree', 'Structure_Numeric': 1, 'duration_numeric': 3, 'height_numeric': 3},
    'Jowar': {'Duration': 'Low', 'Height': 'Medium', 'Structure': 'Grass Crop', 'Structure_Numeric': 3, 'duration_numeric': 1, 'height_numeric': 2},
    'Lemon': {'Duration': 'Long', 'Height': 'Tall', 'Structure': 'Tree', 'Structure_Numeric': 1, 'duration_numeric': 3, 'height_numeric': 3},
    'Maize': {'Duration': 'Medium', 'Height': 'Medium', 'Structure': 'Grass Crop', 'Structure_Numeric': 3, 'duration_numeric': 2, 'height_numeric': 2},
    'Mangoes': {'Duration': 'Long', 'Height': 'Tall', 'Structure': 'Tree', 'Structure_Numeric': 1, 'duration_numeric': 3, 'height_numeric': 3},
    'Paddy': {'Duration': 'Medium', 'Height': 'Short', 'Structure': 'Grass Crop', 'Structure_Numeric': 3, 'duration_numeric': 2, 'height_numeric': 1},
    'Ragi': {'Duration': 'Low', 'Height': 'Short', 'Structure': 'Grass Crop', 'Structure_Numeric': 3, 'duration_numeric': 1, 'height_numeric': 1},
    'Redgram': {'Duration': 'Medium', 'Height': 'Medium', 'Structure': 'Shrub', 'Structure_Numeric': 2, 'duration_numeric': 2, 'height_numeric': 2},
    'Rose': {'Duration': 'Medium', 'Height': 'Short', 'Structure': 'Shrub', 'Structure_Numeric': 2, 'duration_numeric': 2, 'height_numeric': 1},
    'Sapota': {'Duration': 'Long', 'Height': 'Tall', 'Structure': 'Tree', 'Structure_Numeric': 1, 'duration_numeric': 3, 'height_numeric': 3},
    'Sugarcane': {'Duration': 'Long', 'Height': 'Tall', 'Structure': 'Grass Crop', 'Structure_Numeric': 3, 'duration_numeric': 3, 'height_numeric': 3}
}

def append_crop_attributes_inplace(file_path):
    df = pd.read_csv(file_path)

    # Initialize new columns if they don't exist
    for col in ["Height", "height_numeric", "Structure", "Structure_Numeric", "Duration", "duration_numeric"]:
        if col not in df.columns:
            df[col] = None

    # Fill values based on Crop_Name
    for i, row in df.iterrows():
        crop = row.get("Crop_Name")
        if crop in crop_mapping:
            for key, value in crop_mapping[crop].items():
                df.at[i, key] = value

    # Save back to same file
    df.to_csv(file_path, index=False)
    print(f"Updated and saved: {file_path}")
```

```python
append_crop_attributes_inplace("Karnataka_Datasets/Across/Kharif/Davanagere/Davanagere_Kharif_Data_Final.csv")
```
