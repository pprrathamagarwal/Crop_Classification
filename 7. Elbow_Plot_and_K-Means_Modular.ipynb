# All Imports

```python
from sklearn.cluster import KMeans
from sklearn.metrics import pairwise_distances_argmin_min
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from tslearn.clustering import TimeSeriesKMeans
from tslearn.preprocessing import TimeSeriesScalerMeanVariance
import matplotlib.pyplot as plt
import numpy as np
import os
import pandas as pd
import re
```

# Elbow Plot

```python

# Load the interpolated CSV
df = pd.read_csv("Karnataka_Datasets/Across/Kharif/Davanagere/Davanagere_SAR_NDVI_Interpolated.csv")

# Identify GCVI columns (e.g., GCVI_1 to GCVI_37)
gcvi_columns = sorted([col for col in df.columns if re.match(r'GCVI_\d+$', col)],
                      key=lambda x: int(x.split('_')[1]))

# Group by Crop_Name
crop_groups = df.groupby('Crop_Name')

for crop, group in crop_groups:
    gcvi_data = group[gcvi_columns]

    # Skip groups with too few samples
    if len(gcvi_data) < 2:
        print(f"Skipping {crop} — not enough samples ({len(gcvi_data)})")
        continue

    # Scale the data
    scaler = StandardScaler()
    data_scaled = scaler.fit_transform(gcvi_data)

    # Compute distortions for different K values (up to number of samples)
    distortions = []
    max_k = min(10, len(gcvi_data))  # Avoid k > number of samples
    K = range(1, max_k + 1)

    for k in K:
        kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
        kmeans.fit(data_scaled)
        distortions.append(kmeans.inertia_)

    # Plot the elbow curve
    plt.figure()
    plt.plot(K, distortions, 'bo-')
    plt.title(f'Elbow Method for Crop: {crop} (n={len(gcvi_data)})')
    plt.xlabel('Number of Clusters (K)')
    plt.ylabel('Distortion (Inertia)')
    plt.grid(True)
    plt.tight_layout()
    plt.show()
```

# K Means Clustering

```python

# Load data
df = pd.read_csv("Karnataka_Datasets/Across/Kharif/Davanagere/Davanagere_SAR_NDVI_Interpolated.csv")

# Extract GCVI time series columns
gcvi_columns = sorted([col for col in df.columns if re.match(r'GCVI_\d+$', col)],
                      key=lambda x: int(x.split('_')[1]))

# Define number of clusters per crop
k_per_crop = {
    'Arecanut': 8,
    'Avare': 4,
    'Bajra': 4,
    'Banana': 3,
    'Black Pepper': 4,
    'Eucalyptus': 1,
    'Coconut': 4,
    'Grapes': 1,
    'Guava': 1,
    'Jowar': 4,
    'Lemon': 1,
    'Maize': 5,
    'Mangoes': 4,
    'Onion': 5,
    'Paddy': 5,
    'Potato': 6,
    'Ragi': 5,
    'Redgram': 5,
    'Rose': 1,
    'Sapota': 1,
    'Sugarcane': 5
    
    
    # Add more if needed
}

# Output dirs
os.makedirs("Karnataka_Datasets/Across/Kharif/Davanagere/Davanagere_Clusters", exist_ok=True)
os.makedirs("Karnataka_Datasets/Across/Kharif/Davanagere/cluster_plots", exist_ok=True)

# Group by crop
crop_groups = df.groupby("Crop_Name")

for crop, group in crop_groups:
    if crop not in k_per_crop:
        continue

    print(f"Processing crop: {crop}")
    k = k_per_crop[crop]
    group_clean = group.dropna(subset=gcvi_columns).copy()

    if group_clean.empty:
        print(f"Skipping {crop} due to no clean data.")
        continue

    # Prepare time series data
    ts_data = group_clean[gcvi_columns].to_numpy()
    ts_scaled = TimeSeriesScalerMeanVariance().fit_transform(ts_data)

    # TimeSeries KMeans
    model = TimeSeriesKMeans(n_clusters=k, metric="dtw", random_state=42)
    labels = model.fit_predict(ts_scaled)

    # Closest to centroid
    closest_indices, _ = pairwise_distances_argmin_min(model.cluster_centers_.reshape(k, -1),
                                                        ts_scaled.reshape(len(ts_scaled), -1))

    # Add clustering info
    group_clean["Cluster"] = labels
    group_clean["IsCentroid"] = False
    for idx in closest_indices:
        group_clean.iloc[idx, group_clean.columns.get_loc("IsCentroid")] = True

    # Save cluster CSV
    csv_path = f"Karnataka_Datasets/Across/Kharif/Davanagere/Davanagere_Clusters/{crop}_clusters.csv"
    group_clean.to_csv(csv_path, index=False)
    print(f"Saved CSV for {crop}: {csv_path}")

    # --- Visualization ---
    plt.figure(figsize=(3 * k, 5))
    for cluster_id in range(k):
        plt.subplot(1, k, cluster_id + 1)

        cluster_series = ts_scaled[labels == cluster_id]

        for ts in cluster_series[:10]:  # plot up to 10
            plt.plot(ts.ravel(), color="gray", alpha=0.4)

        plt.plot(model.cluster_centers_[cluster_id].ravel(), color="red", linewidth=2, label="Centroid")
        plt.title(f"{crop} - Cluster {cluster_id + 1}")
        plt.ylim(np.nanmin(ts_scaled), np.nanmax(ts_scaled))
        plt.grid(True)
        plt.tight_layout()

    plt.suptitle(f"GCVI Clustering for {crop} (K={k})", fontsize=14)
    plot_path = f"Karnataka_Datasets/Across/Kharif/Davanagere/cluster_plots/{crop}_clusters.png"
    plt.savefig(plot_path)
    plt.show()
    print(f"Saved plot for {crop}: {plot_path}")
```

# Computing Distances

```python

# Directories
input_dir = "Karnataka_Datasets/Across/Kharif/Davanagere/Davanagere_Clusters/"
output_dir = "Karnataka_Datasets/Across/Kharif/Davanagere/Davanagere_Clusters_With_Distance/"
os.makedirs(output_dir, exist_ok=True)

# Process all CSV files in the input directory
for filename in os.listdir(input_dir):
    if not filename.endswith(".csv"):
        continue

    input_path = os.path.join(input_dir, filename)
    output_path = os.path.join(output_dir, filename)

    print(f"Processing: {filename}")
    df = pd.read_csv(input_path)

    # Extract GCVI columns and sort them
    gcvi_columns = sorted(
        [col for col in df.columns if re.match(r'GCVI_\d+$', col)],
        key=lambda x: int(x.split('_')[1])
    )

    if "Cluster" not in df.columns:
        print(f"Skipping {filename}: 'Cluster' column missing.")
        continue

    # Ensure Distance column exists
    df["Distance"] = np.nan

    # Loop through each cluster
    for cluster_id in sorted(df["Cluster"].unique()):
        cluster_data = df[df["Cluster"] == cluster_id]

        # Skip if cluster is empty or only has one point
        if cluster_data.shape[0] <= 1:
            print(f"Skipping cluster {cluster_id} in {filename}: not enough points.")
            continue

        # Determine centroid
        if "IsCentroid" in df.columns and any(cluster_data["IsCentroid"] == True):
            centroid_vector = cluster_data[cluster_data["IsCentroid"] == True].iloc[0][gcvi_columns].to_numpy()
            print(f"Cluster {cluster_id} in {filename}: Used marked centroid.")
        else:
            centroid_vector = cluster_data[gcvi_columns].median().to_numpy()
            print(f"Cluster {cluster_id} in {filename}: Computed centroid manually.")

        # Compute distances for this cluster
        for idx in cluster_data.index:
            ts_vector = df.loc[idx, gcvi_columns].to_numpy()
            distance = np.linalg.norm(ts_vector - centroid_vector)
            df.at[idx, "Distance"] = distance

    # Save the modified file
    df.to_csv(output_path, index=False)
    print(f"Saved: {output_path}\n")
```

# Outlier Removing

```python
# import pandas as pd

# Input directory with distances
input_dir = "Karnataka_Datasets/Across/Kharif/Davanagere/Davanagere_Clusters_With_Distance/"

# Output directories
outlier_dir = "Karnataka_Datasets/Across/Kharif/Davanagere/clustered_crops_outliers"
filtered_dir = "Karnataka_Datasets/Across/Kharif/Davanagere/clustered_crops_filtered"

os.makedirs(outlier_dir, exist_ok=True)
os.makedirs(filtered_dir, exist_ok=True)

for filename in os.listdir(input_dir):
    if not filename.endswith(".csv"):
        continue

    print(f"Processing: {filename}")
    input_path = os.path.join(input_dir, filename)
    df = pd.read_csv(input_path)

    if "Cluster" not in df.columns or "Distance" not in df.columns:
        print(f"Skipping {filename}: Missing 'Cluster' or 'Distance' column.")
        continue

    # DataFrames to collect final results
    all_outliers = []
    all_filtered = []

    for cluster_id in sorted(df["Cluster"].unique()):
        cluster_data = df[df["Cluster"] == cluster_id]

        if cluster_data.shape[0] < 4:
            # Skip small clusters
            all_filtered.append(cluster_data)
            continue

        q75 = cluster_data["Distance"].quantile(0.75)

        outliers = cluster_data[cluster_data["Distance"] > q75]
        remaining = cluster_data[cluster_data["Distance"] <= q75]

        all_outliers.append(outliers)
        all_filtered.append(remaining)

    # Concatenate and save
    outlier_df = pd.concat(all_outliers, ignore_index=True)
    filtered_df = pd.concat(all_filtered, ignore_index=True)

    outlier_df.to_csv(os.path.join(outlier_dir, filename), index=False)
    filtered_df.to_csv(os.path.join(filtered_dir, filename), index=False)

    print(f"Saved outliers → {os.path.join(outlier_dir, filename)}")
    print(f"Saved filtered → {os.path.join(filtered_dir, filename)}\n")
```

# Removing Duplicates

```python

# Load your CSV
df = pd.read_csv("Karnataka_Datasets/Across/S2/Merged_Karnataka_S2_Kharif_Mapped.csv")

# Group VI columns by base name (e.g., NDVI, GCVI)
vi_groups = {}
for col in df.columns:
    match = re.match(r"([A-Z]+VI)_\d+", col)
    if match:
        base = match.group(1)
        vi_groups.setdefault(base, []).append(col)

# Keep track of all rows to keep
mask = pd.Series([True] * len(df))

# For each VI group, remove duplicate rows within that group
for group_cols in vi_groups.values():
    group_dupes = df.duplicated(subset=group_cols)
    mask &= ~group_dupes  # Only keep the first occurrence of duplicates

# Filter the DataFrame
deduplicated_df = df[mask].reset_index(drop=True)

# Count how many rows were dropped
num_dropped = len(df) - len(deduplicated_df)
print(f"Rows dropped due to VI duplicates: {num_dropped}")
# Save the cleaned data if needed
deduplicated_df.to_csv("Karnataka_Datasets/Across/S2/Merged_Karnataka_S2_Kharif_Mapped_Clean.csv", index=False)
```
