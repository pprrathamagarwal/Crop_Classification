# Printing Unique Crop Names

```python


# Load the CSV file
file_path = 'Karnataka_Datasets/Across/Clean/SAR_Mapped.csv'  # <-- Replace with your CSV file path
df = pd.read_csv(file_path)

# Check if 'Crop_Name' column exists
if 'Crop_Name' in df.columns:
    # Get unique values
    unique_crops = df['Crop_Name'].dropna().unique()
    
    # Print each unique crop
    print("Unique Crop Names:")
    for crop in unique_crops:
        print(crop)
else:
    print("'Crop_Name' column not found in the file.")
```

# Keeping Only A Set Of Crops With Their Height and Duration Mappings

```python
# Allowed crop names
allowed_crops = [
    "Mangoes", "Paddy", "Ragi", "Redgram", "Rose", "Sapota", "Sugarcane",
    "Maize", "Arecanut", "Avare", "Bajra", "Banana", "Coconut",
    "Eucalyptus", "Guava", "Jowar", "Lemon"
]

height_mapping = {
    "Arecanut": "Tall",
    "Avare": "Medium",
    "Bajra": "Short",
    "Banana": "Tall",
    "Coconut": "Tall",
    "Eucalyptus": "Tall",
    "Guava": "Tall",
    "Jowar": "Medium",
    "Lemon": "Tall",
    "Maize": "Medium",
    "Mangoes": "Tall",
    "Paddy": "Short",
    "Ragi": "Short",
    "Redgram": "Medium",
    "Rose": "Short",
    "Sapota": "Tall",
    "Sugarcane": "Tall"
}

height_numeric_mapping = {
    "Short": 1,
    "Medium": 2,
    "Tall": 3
}

duration_mapping = {
    "Arecanut": "Long",
    "Avare": "Low",
    "Bajra": "Low",
    "Banana": "Long",
    "Coconut": "Long",
    "Eucalyptus": "Long",
    "Guava": "Long",
    "Jowar": "Low",
    "Lemon": "Long",
    "Maize": "Medium",
    "Mangoes": "Long",
    "Paddy": "Medium",
    "Ragi": "Low",
    "Redgram": "Medium",
    "Rose": "Medium",
    "Sapota": "Long",
    "Sugarcane": "Long"
}

duration_numeric_mapping = {
    "Low": 1,
    "Medium": 2,
    "Long": 3
}

# Load your CSV
file_path = 'Karnataka_Datasets/Across/Clean/SAR_Mapped.csv'  # <-- Replace with your filename
df = pd.read_csv(file_path)

# Step 1: Filter only allowed crops
df = df[df['Crop_Name'].isin(allowed_crops)].copy()

# Step 2: Map the height and duration columns
df['height'] = df['Crop_Name'].map(height_mapping)
df['height_numeric'] = df['height'].map(height_numeric_mapping)
df['duration'] = df['Crop_Name'].map(duration_mapping)
df['duration_numeric'] = df['duration'].map(duration_numeric_mapping)

# Save or display
print(df)

# Optionally save to new CSV
df.to_csv('Karnataka_Datasets/Across/Clean/SAR_Mapped_Height_Duration_Structure.csv', index=False)
```

```python


# Load your data
df = pd.read_csv('Karnataka_Datasets/Across/Clean/SAR_Mapped_Height_Duration_Structure.csv')  # replace with your actual filename

# Separate Ragi and other crops
ragi_df = df[df['Crop_Name'] == 'Ragi']
other_df = df[df['Crop_Name'] != 'Ragi']

# Randomly sample 7000 Ragi samples
ragi_downsampled = ragi_df.sample(n=7000, random_state=42)

# Combine back
final_df = pd.concat([ragi_downsampled, other_df]).reset_index(drop=True)

# Check the new Crop_Name distribution
print(final_df['Crop_Name'].value_counts())

# Optionally save it
final_df.to_csv('Karnataka_Datasets/Across/Clean/SAR_Mapped_Height_Duration_Structure_Clean.csv', index=False)
```

```python


# Load your dataset
df = pd.read_csv('Karnataka_Datasets/Across/Clean/SAR_Mapped_Height_Duration_Structure_Clean.csv')  # Replace with your filename

# Create empty lists to hold train and test data
train_list = []
test_list = []

# Group by Crop_Name and split each group
for crop, group in df.groupby('Crop_Name'):
    train_crop = group.sample(frac=0.8, random_state=42)
    test_crop = group.drop(train_crop.index)

    train_list.append(train_crop)
    test_list.append(test_crop)

# Concatenate all crops' splits
train_df = pd.concat(train_list).reset_index(drop=True)
test_df = pd.concat(test_list).reset_index(drop=True)

# Check the distribution
print('Training set Crop_Name distribution:')
print(train_df['Crop_Name'].value_counts())

print('\nTesting set Crop_Name distribution:')
print(test_df['Crop_Name'].value_counts())

# Optionally save to files
train_df.to_csv('Karnataka_Datasets/Across/Clean/Bangalore_train_height_duration_structure_train.csv', index=False)
test_df.to_csv('Karnataka_Datasets/Across/Clean/Bangalore_train_height_duration_structure_test.csv', index=False)
```
