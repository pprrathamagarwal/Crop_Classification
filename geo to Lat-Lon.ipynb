```python
import pandas as pd
import os
import json

# Path to your directory containing the CSV files
directory = "Karnataka_Datasets/Across/Kharif/Cropland_Masked///"

# Iterate over all CSV files in the directory
for filename in os.listdir(directory):
    if filename.endswith(".csv"):
        filepath = os.path.join(directory, filename)
        try:
            df = pd.read_csv(filepath)

            # Proceed only if '.geo' column exists
            if '.geo' in df.columns:
                # Parse coordinates from the .geo column
                def extract_lat_lon(geo_str):
                    try:
                        geo_json = json.loads(geo_str)
                        lon, lat = geo_json["coordinates"]
                        return pd.Series([lat, lon])
                    except Exception:
                        return pd.Series([None, None])

                df[['Latitude', 'Longitude']] = df['.geo'].apply(extract_lat_lon)

                # Optionally drop the '.geo' column
                df.drop(columns=['.geo'], inplace=True)

                # Save the updated file in-place
                df.to_csv(filepath, index=False)
                print(f"✅ Updated: {filename}")
            else:
                print(f"⏭️ Skipped (no .geo column): {filename}")

        except Exception as e:
            print(f"❌ Failed: {filename} due to {e}")
```
