```python
import os
import pandas as pd

def remove_files_without_column(directory, required_column):
    """
    Removes CSV files in the specified directory if the required column is missing.

    :param directory: Path to the directory containing CSV files.
    :param required_column: The column that must exist in each CSV file.
    """
    # Ensure the directory exists
    if not os.path.exists(directory):
        print(f"Directory '{directory}' does not exist.")
        return

    # Iterate through each file in the directory
    for filename in os.listdir(directory):
        if filename.endswith('.csv'):
            file_path = os.path.join(directory, filename)
            try:
                # Load the CSV file
                df = pd.read_csv(file_path)
                # Check if the required column exists
                if required_column not in df.columns:
                    print(f"Removing {filename} - Missing column '{required_column}'")
                    os.remove(file_path)
                else:
                    print(f"{filename} - Column '{required_column}' exists")
            except Exception as e:
                print(f"Error processing {filename}: {e}")
                # Optionally, handle corrupt files here if needed

# Example usage
directory_path = "Karnataka_Datasets/Atibelle_Extracted/Atibelle_Data/"  # Replace with your directory path
required_column_name = "ARVI"  # Replace with the column you want to check
remove_files_without_column(directory_path, required_column_name)
```

```python
import os
import pandas as pd
from glob import glob

def merge_and_sort_files(directory_path, output_file):
    # Get all CSV files in the directory
    all_files = glob(os.path.join(directory_path, '*.csv'))
    
    # Sort files by their names to ensure the correct order
    all_files.sort()
    
    # Initialize an empty DataFrame for appending rows
    merged_df = pd.DataFrame()
    
    # Iterate over each file and append rows based on (latitude, longitude)
    for file_path in all_files:
        df = pd.read_csv(file_path)
        
        # Append the rows from the current file to the merged DataFrame
        merged_df = pd.concat([merged_df, df])
    
    # Sort the merged DataFrame by latitude and longitude
    merged_df = merged_df.sort_values(by=['Latitude', 'Longitude']).reset_index(drop=True)
    
    # Save the merged and sorted DataFrame to a CSV file
    merged_df.to_csv(output_file, index=False)
    print(f'Saved merged and sorted file: {output_file}')

# Example usage
directory_path = r"Karnataka_Datasets/District_Data_NDVI_5Day/Merged/"
output_file = r"Karnataka_Datasets/District_Data_NDVI_5Day/Merged/Karnataka_Complete_Data.csv"
merge_and_sort_files(directory_path, output_file)
```

```python
import os
import pandas as pd

# Input CSV file path
input_file = 'Karnataka_Datasets/Model_Data/Karnataka_test.csv'

# Directory to save the chunked CSV files
output_directory = 'Karnataka_Datasets/Model_Data/Test_Chunks/'

# Create the output directory if it doesn't exist
if not os.path.exists(output_directory):
    os.makedirs(output_directory)

# Read the CSV file into a DataFrame
df = pd.read_csv(input_file)

# Determine the number of chunks (each chunk will contain 100 rows)
chunk_size = 380
total_rows = len(df)
num_chunks = (total_rows // chunk_size) + (1 if total_rows % chunk_size != 0 else 0)

# Loop to save chunks
for i in range(num_chunks):
    # Calculate the start and end indices for each chunk
    start_idx = i * chunk_size
    end_idx = min((i + 1) * chunk_size, total_rows)
    
    # Slice the DataFrame to get the chunk
    chunk = df.iloc[start_idx:end_idx]
    
    # Define the output file name for the chunk
    chunk_filename = f"chunk_{i+1}.csv"
    chunk_filepath = os.path.join(output_directory, chunk_filename)
    
    # Save the chunk as a CSV file
    chunk.to_csv(chunk_filepath, index=False)

    print(f"Chunk {i+1} saved as {chunk_filename}")

print("CSV file split into chunks successfully.")
```

```python
import os
import pandas as pd
from sklearn.model_selection import train_test_split

# Define the directory containing the CSV files
input_directory = "Karnataka_Datasets/Model_Data/"
output_train_file = "Karnataka_Datasets/Model_Data/Karnataka_train.csv"
output_test_file = "Karnataka_Datasets/Model_Data/Karnataka_test.csv"

# Merge all CSV files into one DataFrame
all_files = [os.path.join(input_directory, f) for f in os.listdir(input_directory) if f.endswith('.csv')]
df = pd.concat((pd.read_csv(file) for file in all_files), ignore_index=True)

# Check for duplicates before removing them
num_duplicates = df.duplicated().sum()
print(f"Number of duplicate rows found: {num_duplicates}")

# Remove duplicate rows
df = df.drop_duplicates()

# Perform an 80-20 split based on "Crop_Name"
train_list = []
test_list = []

for crop_name, group in df.groupby("Crop_Name"):
    train, test = train_test_split(group, test_size=0.2, random_state=42)
    train_list.append(train)
    test_list.append(test)

# Concatenate the splits
train_df = pd.concat(train_list, ignore_index=True)
test_df = pd.concat(test_list, ignore_index=True)

# Save the splits into CSV files
train_df.to_csv(output_train_file, index=False)
test_df.to_csv(output_test_file, index=False)

print(f"Train set saved to {output_train_file}")
print(f"Test set saved to {output_test_file}")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the CSV file
file_path = "Karnataka_Datasets/Model_Data/final_train_height.csv"  # Replace with your actual file path
df = pd.read_csv(file_path)

# Extract columns for VV and VH values
vv_columns = [col for col in df.columns if col.startswith("VV_")]
vh_columns = [col for col in df.columns if col.startswith("VH_")]

# Extract max VV and max VH across all time steps for each row
df["max_VV"] = df[vv_columns].max(axis=1)
df["max_VH"] = df[vh_columns].max(axis=1)

# Calculate the VV/VH ratio (avoid division by zero)
df["VV_VH_ratio"] = df["max_VV"] / df["max_VH"]

# Handle any division by zero or invalid values by replacing them with NaN
df["VV_VH_ratio"] = df["VV_VH_ratio"].replace([float('inf'), -float('inf')], float('nan'))

# Group by 'Name' and 'Crop_Name'
grouped = df.groupby("Name")

# Create boxplots for each 'Crop_Name' within each 'Name'
for name, group in grouped:
    plt.figure(figsize=(10, 6))
    sns.boxplot(
        data=group,
        x="Crop_Name",
        y="VV_VH_ratio",
        hue="Crop_Name",
        showfliers=False,  # Optional: hide outliers
        palette="Set2",
    )
    plt.title(f"Boxplot of VV/VH ratio for Name: {name}")
    plt.ylabel("VV/VH Ratio")
    plt.xlabel("Crop Name")
    plt.legend(title="Crop Name", loc="best")
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()
```

```python
import pandas as pd
import numpy as np

# Load CSV files
file1 = "Code-20250212T204009Z-001/Code/cluster_centroids_with_mapping.csv"  # First file (contains class labels and cluster info)
file2 = "Code-20250212T204009Z-001/Code/classified_output.csv"  # Second file (contains NDVI data)

df1 = pd.read_csv(file1)
df2 = pd.read_csv(file2)

# Extract relevant columns
timesteps = [f"TimeStep_{i}" for i in range(12)]
ndvi_steps = [f"{i}_gapfilled_NDVI_lsc" for i in range(12)]

X1 = df1[timesteps].values  # Matrix from first file
X2 = df2[ndvi_steps].values  # Matrix from second file

# Compute Euclidean distance for each row in df2 against all rows in df1
nearest_classes = []
nearest_clusters = []  # To store cluster numbers

for idx2, row2 in enumerate(X2):
    distances = np.linalg.norm(X1 - row2, axis=1)  # Compute Euclidean distance
    nearest_idx = np.argmin(distances)  # Index of the closest row
    
    # Get the corresponding class and cluster
    nearest_class = df1.iloc[nearest_idx]["class"]
    nearest_cluster = df1.iloc[nearest_idx]["cluster"]
    
    # Append values to lists
    nearest_classes.append(nearest_class)
    nearest_clusters.append(nearest_cluster)

    # Debugging: Print the closest match found
    print(f"Row {idx2}: Closest match is row {nearest_idx} with distance {distances[nearest_idx]:.4f}, "
          f"assigned class: {nearest_class}, assigned cluster: {nearest_cluster}")

# Add the matched class and cluster columns to df2
df2["class"] = nearest_classes
df2["cluster"] = nearest_clusters

# Save the output
output_file = "Code-20250212T204009Z-001/Code/classified_output.csv"
df2.to_csv(output_file, index=False)

print(f"Processing complete. Output saved as '{output_file}'.")
```

```python
import pandas as pd

# Load the CSV file
file_path = "Code-20250212T204009Z-001/Code/classified_output.csv"
df = pd.read_csv(file_path)

# Define the dictionary with class sets for each duration category
duration_mapping = {
    "low": {13, 15, 16, 17, 20},
    "medium": {14, 16, 17, 18, 19,21},
    "high": {19, 20,21}
}

# Initialize predicted column with 0
df["predicted"] = 0  

# Function to determine predicted class
def determine_predicted(row):
    class_val = row["class"]
    actual_val = row["actual"]

    # Ambiguous classes with priority on actual value
    if class_val in {20, 19, 17, 16, 21}:
        if actual_val == 3 and class_val == 21:
            return 3
        if actual_val == 3 and class_val == 20:
            return 3  # High
        if actual_val == 2 and class_val in {19, 17, 16}:
            return 2  # Medium
        if actual_val == 1 and class_val in {20, 17, 16}:
            return 1  # Low
    
    # Normal mapping
    for duration, class_set in duration_mapping.items():
        if class_val in class_set:
            return {"low": 1, "medium": 2, "high": 3}[duration]

    return 0  # Default (if none match)

# Apply function to determine predictions
df["predicted"] = df.apply(determine_predicted, axis=1)

# Save the updated CSV file
output_path = "Code-20250212T204009Z-001/Code/classified_output_.csv"
df.to_csv(output_path, index=False)

# Print preview
print(f"Updated file saved as '{output_path}'.")
print(df.head())
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix
from tabulate import tabulate

# Load the CSV file
file_path = "Code-20250212T204009Z-001/Code/classified_output_.csv"
df = pd.read_csv(file_path)

# Ensure "actual" and "predicted" are integers
df["actual"] = df["actual"].astype(int)
df["predicted"] = df["predicted"].astype(int)

# Extract true and predicted labels
y_true = df["actual"]
y_pred = df["predicted"]

#  Calculate direct metrics
accuracy = accuracy_score(y_true, y_pred)
precision = precision_score(y_true, y_pred, average=None, zero_division=0)
recall = recall_score(y_true, y_pred, average=None, zero_division=0)
f1 = f1_score(y_true, y_pred, average=None, zero_division=0)

#  Calculate overall metrics (weighted average)
precision_overall = precision_score(y_true, y_pred, average='weighted', zero_division=0)
recall_overall = recall_score(y_true, y_pred, average='weighted', zero_division=0)
f1_overall = f1_score(y_true, y_pred, average='weighted', zero_division=0)

#  Identify unique class labels
classes = sorted(set(y_true.unique()).union(set(y_pred.unique())))
class_labels = [str(c) for c in classes]

#  Create table for overall metrics
overall_table = [
    ["Metric", "Value"],
    ["Accuracy", f"{accuracy:.4f}"],
    ["Precision", f"{precision_overall:.4f}"],
    ["Recall", f"{recall_overall:.4f}"],
    ["F1-Score", f"{f1_overall:.4f}"]
]

print(" Overall Performance:")
print(tabulate(overall_table, headers="firstrow", tablefmt="grid"))
print("\n")

#  Create table for per-class metrics
class_table = [
    ["Class"] + class_labels,
    ["Precision"] + [f"{p:.4f}" for p in precision],
    ["Recall"] + [f"{r:.4f}" for r in recall],
    ["F1-Score"] + [f"{f:.4f}" for f in f1],
    ["Support"] + [str(sum(y_true == int(c))) for c in class_labels]
]

print(" Per-Class Performance:")
print(tabulate(class_table, headers="firstrow", tablefmt="grid"))
print("\n")

#  Compute Confusion Matrix
cm = confusion_matrix(y_true, y_pred, labels=classes)

#  Print Confusion Matrix in tabular format
cm_table = [[label] + [str(count) for count in row] for label, row in zip(class_labels, cm)]
cm_table.insert(0, ["True \\ Pred"] + class_labels)

print(" Confusion Matrix:")
print(tabulate(cm_table, headers="firstrow", tablefmt="grid"))

#  Plot the Confusion Matrix
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt="d", cmap="Blues", xticklabels=class_labels, yticklabels=class_labels)
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.title("Confusion Matrix")
plt.show()
```

```python
import pandas as pd
import numpy as np
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

# Load train and test CSV files
train_df = pd.read_csv("Karnataka_Datasets/Model_Data/final_train_height.csv")
test_df = pd.read_csv("Karnataka_Datasets/Bangalore/")

# Identify unique time steps (extracting numbers before _VV/_VH)
time_steps = sorted(set(int(col.split('_')[1]) for col in train_df.columns if '_VV' in col))

# Identify feature columns
vv_vh_columns = [f"VV_{t}" for t in time_steps] + [f"{VH_{t}" for t in time_steps]

# Target column
target_col = "duration_numeric"

# Ensure the dataset contains all required columns
missing_columns = [col for col in vv_vh_columns if col not in train_df.columns]
if missing_columns:
    raise ValueError(f"Missing expected columns: {missing_columns}")

# Extract features and target
X_train = train_df[vv_vh_columns].values  # Convert to numpy array
y_train = train_df[target_col]

X_test = test_df[vv_vh_columns].values
y_test = test_df[target_col]

# Train a Random Forest Classifier
rf = RandomForestClassifier(n_estimators=100, random_state=42)
rf.fit(X_train, y_train)

# Predict on test data
y_pred = rf.predict(X_test)

# Evaluate model performance
accuracy = accuracy_score(y_test, y_pred)
print(f"Test Accuracy: {accuracy:.4f}")

# Save predictions
test_df["predicted_duration_numeric"] = y_pred
test_df.to_csv("Karnataka_Datasets/Model_Data/test_with_predictions.csv", index=False)
```

# Remove Unncessary Crops

```python
import pandas as pd

# === Load the CSV ===
input_file = "Karnataka_Datasets/Rural/Bangalore_Rural_cleaned.csv"  # Replace with your input file name
output_file = "Karnataka_Datasets/Rural/Bangalore_Rural_cleaned.csv"

# Read CSV
df = pd.read_csv(input_file)

# === Clean the Crop Names ===
crop_column = "Cropname"  # Update if your column has a different name

# 1. Standardize Crop Names
df[crop_column] = df[crop_column].str.strip()  # Remove leading/trailing spaces
df[crop_column] = df[crop_column].str.replace(r'-(H|L)$', '', regex=True)  # Remove '-H' and '-L'
df[crop_column] = df[crop_column].str.replace(r'\s+', ' ', regex=True)  # Collapse multiple spaces
df[crop_column] = df[crop_column].str.replace(r'\s*:\s*', ': ', regex=True)  # Fix spaces around colons
df[crop_column] = df[crop_column].str.title()  # Standardize capitalization if needed

# 2. Fix known duplicates
fix_duplicates = {
    "Ragi :": "Ragi",
    "Na Land": "NA Land",  # Handle variations
    "Betel Nuts (Areca Nuts)": "Betel Nuts (Areca nuts)",
}
df[crop_column] = df[crop_column].replace(fix_duplicates)

# === Remove Unwanted Crops ===
unwanted_crops = [
    "NA Land", "Fallow", "Grass", "Harvest Over", "Unknown", "Other"
]
df = df[~df[crop_column].isin(unwanted_crops)]
df = df[~df[crop_column].str.startswith("Harvest over", na=False)]

# === Group by Crop Name and Count ===
crop_counts = df[crop_column].value_counts().sort_values(ascending=False)

# === Save the Cleaned CSV ===
df.to_csv(output_file, index=False)

# === Print Results ===
print("Cleaning complete! Saved as:", output_file)
print("Crop Names and Their Counts (Sorted by Count):")
print(crop_counts)

```

```python
import pandas as pd

# Load both CSV files and drop 'Correct' column
file1 = "Karnataka_Datasets/Ramanagara/predictions_height.csv"
file2 = "Karnataka_Datasets/Ramanagara/predictions_duration.csv"
df1 = pd.read_csv(file1).drop(columns=["Correct"], errors="ignore")
df2 = pd.read_csv(file2).drop(columns=["Correct"], errors="ignore")

# Merge to keep all columns from df1 and add Predicted_duration from df2
merged_df = pd.merge(df1, df2[["latitude", "longitude", "Predicted_duration"]], 
                     on=["latitude", "longitude"], 
                     how="left")

# Save the merged data to a new CSV
output_file = "Karnataka_Datasets/Ramanagara/pairwise_analysis.csv"
merged_df.to_csv(output_file, index=False)

print(f"Merged data saved to {output_file}")
```

# Merge SAR and NDVI Series

```python
import pandas as pd

# Load the two CSV files
df1 = pd.read_csv("Katihar/Katihar_ND.csv")
df2 = pd.read_csv("Karnataka_Datasets/District_Data_NDVI_5Day/SAR/Bangalore_SAR_flattened.csv")

# Merge the files on latitude and longitude
merged_df = pd.merge(df1, df2, on=["Latitude", "Longitude"], how="inner")

# Drop duplicate lat-long pairs (keeping only the first occurrence)
merged_df = merged_df.drop_duplicates(subset=["Latitude", "Longitude"])

# Save the merged file
merged_df.to_csv("Karnataka_Datasets/District_Data_NDVI_5Day/Merged/Bangalore_Merged.csv", index=False)

print("Merging complete. Output saved as 'merged_output.csv'.")
```

```python
import pandas as pd
import ast

# Load CSV file
df = pd.read_csv("Karnataka_Datasets/Agrifield/Agrifield_NDVI.csv")

# Convert NDVI_TimeSeries from string to list
df["NDVI_TimeSeries"] = df["NDVI_TimeSeries"].apply(ast.literal_eval)

# Expand the list into separate columns
ndvi_df = pd.DataFrame(df["NDVI_TimeSeries"].to_list(), columns=[f"NDVI_{i+1}" for i in range(len(df["NDVI_TimeSeries"][0]))])

# Concatenate with original dataframe (excluding NDVI_TimeSeries)
df = pd.concat([df.drop(columns=["NDVI_TimeSeries"]), ndvi_df], axis=1)

# Save the result
df.to_csv("Karnataka_Datasets/Agrifield/Agrifield_NDVI_flat.csv", index=False)
```
